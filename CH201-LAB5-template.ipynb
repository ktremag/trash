```python
import numpy as np
import matplotlib.pyplot as plt

#establishes dimensions and relative locations of subplots (in this case, 3 rows, one column)
fig, (ax1,ax2,ax3) = plt.subplots(3, 1, figsize=(10, 15), sharex=True)
fig.subplots_adjust(hspace=0)

#experimental data
v = np.array([0, 5, 10, 15, 20, 22, 24, 25, 26, 28, 30, 35, 40, 45, 50])
pH = np.array([1, 1.14, 1.3, 1.51, 1.85, 2.08, 2.57, 7, 11.42, 11.89, 12.1, 12.37, 12.52, 12.63, 12.7])

#first derivative
dv = np.array([(v[i+1] + v[i])/2 for i in range(len(v)-1)])
dpH = np.array([(pH[i+1]-pH[i])/(v[i+1]-v[i]) for i in range (len(v)-1)])

#second derivative
ddv = np.array([(dv[i+1] + dv[i])/2 for i in range(len(dv)-1)])
ddpH = np.array([(dpH[i+1]-dpH[i])/(dv[i+1]-dv[i]) for i in range (len(dv)-1)])

eqp = np.where(ddpH == 0)
veqp = ddv[eqp]

print (eqp)
print (veqp)

#graph of titration alone
ax1.plot(v, pH, color='cornflowerblue', marker='o', markerfacecolor='cornflowerblue', markersize=5)
#ax1.set_xlim(0, 55)
ax1.set_ylim(0,13)
ax1.set_yticks(np.arange(2,14,2))
ax1.set_ylabel('pH')
ax1.grid(True)

#graph of 1st derivative of titration 
ax2.plot(dv, dpH, color='green', marker='o', markerfacecolor='green', markersize=5)
#ax2.set_xlim(0, 55)
ax2.set_ylim(-1,5)
ax2.set_yticks(np.arange(0,5,1))
ax2.set_ylabel('$d$pH/$d$V')
ax2.grid(True)

#graph of 2nd derivative of titration 
ax3.plot(ddv, ddpH, color='orange', marker='o', markerfacecolor='orange', markersize=5)
#ax3.set_xlim(0, 55)
ax3.set_ylim(-3,3)
ax3.set_yticks(np.arange(-3,3,1))
ax3.set_xlabel('Volume (mL)')
ax3.set_ylabel('$d^2$pH/$d\mathrm{V}^2$')
ax3.grid(True)

```
