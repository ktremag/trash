```python
import numpy as np
import matplotlib.pyplot as plt

# Provided data
sugar_concentration = np.array([0, 5, 10, 15, 20, 25])
absorbance = np.array([0.000, 0.247, 0.489, 0.748, 0.940, 1.227])

# Fit a linear trendline (degree 1 polynomial) using numpy's polyfit
slope, intercept = np.polyfit(sugar_concentration, absorbance, 1)

# Generate y values for the trendline using the fitted slope and intercept
trendline = slope * sugar_concentration + intercept

# Plotting the standard curve and linear trendline
plt.scatter(sugar_concentration, absorbance, color='maroon')
plt.plot(sugar_concentration, trendline, linestyle='--', color='maroon', label=f'Trendline: y = {slope:.3f}x + {intercept:.3f}') 

# Adding labels and title
plt.xlabel('Sugar Concentration (Î¼moles)')
plt.ylabel('Absorbance (750 nm)')
#plt.title('Standard Curve with Linear Trendline')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```
